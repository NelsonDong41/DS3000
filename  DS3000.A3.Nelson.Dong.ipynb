```python
## 1
import numpy as np

data = np.loadtxt('exam_grades.csv', delimiter = ",", skiprows = 1)

print(data)
print(data.shape)
```

```python
## 2
## there are 5 columns in the dataset and 233 observations
```

```python
## 3
min = data[:, 4].min()
print(min)

max = data[:, 4].max()
print(max)
```

```python
## 4
min_year = data[:,0].min()
max_year = data[:,0].max()


print("time period covers from " + str(min_year) + " to " + str(max_year))
```

```python
## 5
arr = np.empty(0);
count = 0
i = 0
while(i < (data.shape[0] - 1)):
    if(data[i + 1][0] == data[i][0]):
        count += 1
    else :
        arr = np.append(arr, np.array(count))
        count = 0
    i += 1

print(arr)
```

```python
## 6
for line in data:
    print("Year:", line[0], 
          "Min:", line[1:3].min(), 
          "Max:", line[1:3].max(), 
          "Mean:", round(line[1:3].mean(), 3), 
          "Standard Deviation:", round(line[1:3].std(), 3))

```
